```python
#SIR
import numpy as np

beta = 3
gamma = 1

T = 10    
dt = 0.001
nsteps = round(T/dt)
v_S=np.zeros([nsteps+1])
v_I=np.zeros([nsteps+1])
v_R=np.zeros([nsteps+1])
v_S[0] = 0.99
v_I[0] = 0.01

# SIR with mid point method
for t in range(nsteps):
    v_S[t+1]=v_S[t]-dt*beta*(v_S[t] - (dt/2)*beta*v_S[t]*v_I[t]) * (v_I[t]+(dt/2)*(beta*v_S[t]*v_I[t]-gamma*v_I[t]))
    v_I[t+1]=v_I[t]+dt*beta*(v_S[t] - (dt/2)*beta*v_S[t]*v_I[t]) * (v_I[t]+(dt/2)*beta*v_S[t] * v_I[t] - (dt/2)*gamma*v_I[t]) - gamma*dt*(v_I[t]+(dt/2)*(beta*v_S[t]*v_I[t]-gamma*v_I[t]))
    v_R[t+1]=v_R[t]+gamma*dt*(v_I[t]+(dt/2)*beta*v_S[t]*v_I[t] - (dt/2)*gamma*v_I[t])
```

```python
import matplotlib.pyplot as plt

#SIR diagram
import matplotlib.pyplot as plt
trange=[t*dt for t in range(nsteps+1)]
plt.figure()

plt.plot(trange,v_S,label = 'S')
plt.plot(trange,v_I,label = 'I')
plt.plot(trange,v_R,label = 'R')
plt.legend()

plt.xlabel('Time')
plt.ylabel('Population [%]')
plt.savefig('Diagrams/01_SIR.png', bbox_inches='tight') #Ábra mentése
```
